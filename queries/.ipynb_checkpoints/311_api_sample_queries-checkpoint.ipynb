```python
def multi_polygon_query_points(client, MTPLG):
    start_time = time.clock()
    results = client.get("fhrw-4uyv", 
                         limit = 5000000,
                         where = "within_polygon(location, '{0}')".format(MTPLG))
    time_spent = time.clock() - start_time
    num_entries = len(results)
    memory_size = sys.getsizeof(results)/8e+6
    message = 'num_entries: {0}, memory_usage: {1}, time_spent: {2}'.format(num_entries, memory_size, time_spent)
    return(results, message)
```
